```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('Solarize_Light2')
```

```python
data_df = pd.read_csv('data.csv')

ages = data_df['Age']
dev_salaries = data_df['All_Devs']
py_salaries = data_df['Python']
js_salaries = data_df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# compared_to = overall_median
compared_to = dev_salaries 

plt.fill_between(
    ages, 
    py_salaries, 
    compared_to,
    where=(py_salaries>compared_to),
    interpolate=True,
    alpha=0.25,
    label="Above"
)

plt.fill_between(
    ages, 
    py_salaries, 
    compared_to,
    where=(py_salaries<compared_to),
    interpolate=True,
    color='red',
    alpha=0.25,
    label="below"
)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```
